## Section: 1.1 - Define measures that **do not** include the cost information
- (e.g., confusion matrices, accuracy, precision, recall, F-measures, etc.).
- Consider using: from sklearn.metrics import classification_report, confusion_matrix
***

```python
# using classification_report
def measure_with_classification_report(y_true, y_pred):
  print(classification_report(y_true, y_pred, target_names=["satisfied", "neutral or dissatisfied"]))

# using confusion_matrix
def measure_with_confusion_matrix(y_true, y_pred):
  confusion_matrix_result = confusion_matrix(y_true, y_pred, labels=["satisfied", "neutral or dissatisfied"])
  tn, fp, fn, tp = confusion_matrix_result.ravel()

  print('accuracy: {0:.2f}'.format((tp+tn)/(tn + fp + fn + tp)))
  print("precision: {0: .2f}".format(tp/(tp+fp)))
  print("recall: {0: .2f}".format(tp/(tp+fn)))
  print("F-measures: {0: .2f}".format((2*tp)/(2*tp + fp + fn)))
  print("specificity: {0: .2f}".format(tn / (tn + fp)))
  print("type 1 error: {0: .2f}".format(fp / (tn + fp)))
  print("type 2 error: {0: .2f}".format(fn / (fn + tp)))

  return confusion_matrix_result
```

**Discussion**

I define two functions: measure_with_classification_report, measure_with_confusion_matrix

*   measure_with_classification_report can provide precision, recall, f1-score, and support.

*  measure_with_confusion_matrix provides true-negative, false-positive, false-negative, and true-positive. Based on them, accuracy, precision, recall, F-measures, specificity, type 1 error, and type 2 error. Finally, it returns a confusion matrix to be used for cost calculation.

## Section: 1.2 - Define measures that **do** include the cost information
- (e.g., using cost matrices).
- Consider creating a function that takes a confusion matrix and calculates the cost.
***

```python
# calculate cost using confusion matrix
def calculate_cost(conf_matrix):

#                 PREDICTED VALUES
#                      0     1
    cost_matrix = [[0, -75],   # 0
                   [0, 82.5]]   # 1 TRUE VALUES
                   # TP FN
                   # FP TN
    total = 0
    for r in range(0, 2):
        for c in range(0, 2):
            total = total + cost_matrix[r][c] * conf_matrix[r][c]
    return total
```

**Discussion**

I define a function named calculate_cost to calculate the cost of classification. Input is the confusion matrix calculated above.

A table that lists the benefit and cost dollar amounts for a decision on a single customer looks like below.
